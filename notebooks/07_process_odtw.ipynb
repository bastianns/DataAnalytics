```python
odtw_clean = df.drop(
    columns=[
        "id",
        "kode_provinsi",
        "nama_provinsi",
        "satuan"
    ]
)
```

```python
odtw_clean = odtw_clean.rename(
    columns={
        "kode_kabupaten_kota": "kode_wilayah",
        "nama_kabupaten_kota": "kabupaten_kota",
        "jenis_odtw": "kategori_wisata",
        "jumlah_odtw": "jumlah_objek_wisata"
    }
)
```

```python
odtw_clean = odtw_clean[
    [
        "tahun",
        "kode_wilayah",
        "kabupaten_kota",
        "kategori_wisata",
        "jumlah_objek_wisata"
    ]
]
```

```python
odtw_clean.info()
odtw_clean.head()
```

```python
print("Jumlah Baris :", len(odtw_clean))
print("Jumlah Tahun :", odtw_clean["tahun"].nunique())
print("Jumlah Wilayah :", odtw_clean["kabupaten_kota"].nunique())
print("Kategori Wisata :")
print(sorted(odtw_clean["kategori_wisata"].unique()))
```

```python
odtw_clean.to_csv(
    "../data/processed/odtw_clean.csv",
    index=False
)
```
