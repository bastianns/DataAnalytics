Preprocessing Function

```python
import pandas as pd
import re

def preprocess_hotel(filepath):
    
    # Ambil tahun dari nama file
    year = int(re.search(r'(\d{4})', str(filepath)).group(1))

    # Load data
    df = pd.read_csv(filepath)

    # Rename kolom
    df.columns = [
        "jenis_hotel",
        "jan",
        "feb",
        "mar",
        "apr",
        "may",
        "jun",
        "jul",
        "aug",
        "sep",
        "oct",
        "nov",
        "dec",
        "annual"
    ]

    # Buang header tambahan
    df = df.iloc[3:].reset_index(drop=True)

    # Drop annual
    df = df.drop(columns=["annual"])

    # Convert numerik
    month_cols = [
        "jan","feb","mar","apr","may","jun",
        "jul","aug","sep","oct","nov","dec"
    ]

    df[month_cols] = (
    df[month_cols]
    .replace("-", np.nan)
    .astype(float)
)

    # Tambah tahun
    df["tahun"] = year

    return df
```

```python
hotel_2025 = preprocess_hotel(
    "../data/raw/Tingkat Penghunian Kamar Hotel, 2025.csv"
)

display(hotel_2025.head())

hotel_2025.info()
```

```python
from pathlib import Path

hotel_files = sorted(
    Path("../data/raw").glob(
        "Tingkat Penghunian Kamar Hotel, *.csv"
    )
)

hotel_files
```

```python
# Cell 4 (debug)

for file in hotel_files:
    try:
        df = preprocess_hotel(file)
        print(f"SUCCESS: {file.name}")

    except Exception as e:
        print(f"ERROR: {file.name}")
        print(e)
        break
```

```python
all_hotels = []

for file in hotel_files:

    df = preprocess_hotel(file)

    all_hotels.append(df)

hotel_wide = pd.concat(
    all_hotels,
    ignore_index=True
)

print(hotel_wide.shape)

display(hotel_wide.head())

display(hotel_wide.tail())
```

# Data Transformation

```python
hotel_long = pd.melt(
    hotel_wide,
    id_vars=["tahun", "jenis_hotel"],
    value_vars=[
        "jan","feb","mar","apr","may","jun",
        "jul","aug","sep","oct","nov","dec"
    ],
    var_name="bulan",
    value_name="tpk"
)

hotel_long.head()
```

```python
bulan_map = {
    "jan": "Januari",
    "feb": "Februari",
    "mar": "Maret",
    "apr": "April",
    "may": "Mei",
    "jun": "Juni",
    "jul": "Juli",
    "aug": "Agustus",
    "sep": "September",
    "oct": "Oktober",
    "nov": "November",
    "dec": "Desember"
}

hotel_long["bulan"] = hotel_long["bulan"].map(bulan_map)
```
